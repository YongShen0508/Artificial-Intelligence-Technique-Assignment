# 2.0 Data Preprocessing

## 2.1 Split the dataset into training and testing set

```python
#Checking null value
data.isnull().sum()
```

Since there is no null value in the dataset, we can skip handling missing values.

```python
#Separating the output vector(Cardiovascular_risk(y)) from the input matrix
X = data.drop('Cardiovascular_risk(y)', axis = 1)
y = data['Cardiovascular_risk(y)']
```

```python
#Showing the shapes and types of X and y
print ('Shape of original dataset, train:', data.shape)
print ('X: shape=', X.shape, 'type=', type(X))
print ('y: shape=', y.shape, 'type=', type(y))
```

```python
#Split the input matrix into numerical and categorical sets
X_num = X.drop(["Gender", "Family_history", "Alcohol", "Junk_food", "Snack", "Smoking", "Transportation", "TV", "Discipline"], axis=1)
X_cat = X[["Gender", "Family_history", "Alcohol", "Junk_food", "Snack", "Smoking", "Transportation", "TV", "Discipline"]]

print('X_num shape:', X_num.shape)
print('X_cat shape:', X_cat.shape)
```

```python
print('>>> Mean of all columns:\n', X_num.mean(axis=0))
print('\n>>> Std of all columns:\n', X_num.std(axis=0))
```

```python
from sklearn.impute import SimpleImputer
median_imputer = SimpleImputer(strategy = "median")
X_num = median_imputer.fit_transform(X_num)
```

```python
# Your code here (4 lines)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler(copy = False)
scaler.fit(X_num)
X_train_num = scaler.transform(X_num)
```

```python
#Encoding the categorical feature into its one-hot representation
from sklearn.preprocessing import OneHotEncoder

# Using OneHotEncoder to encode multiple categorical features
encoder = OneHotEncoder(sparse_output=False)
X_train_cat = encoder.fit_transform(X_cat)
```

```python
print('Shape of X_train_cat_tr:', X_train_cat.shape, '\n')
print(X_train_cat, '\n')
```

```python
# Create a DataFrame for the transformed data with correct column names
X_train_cat_df = pd.DataFrame(X_train_cat, columns=encoder.get_feature_names_out(X_cat.columns))

# Access and print the possible values for each feature (categories)
for feature, categories in zip(X_train_cat_df.columns, encoder.categories_):
    print(f"Feature '{feature}' possible values: {categories}")
```

These are the possible values for the transformed feature.

```python
#Combining the transformed numerical and categorical sets
X_train_tr = np.hstack([X_train_num, X_train_cat])

print('X_train shape:', X_train_tr.shape)
print('y_train shape:', y.shape)
```

```python
#Split the dataset into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_train_tr, y, test_size=0.2, random_state=30)

print ('full set shape=', X.shape)
print ('train shape=', X_train.shape)
print ('test shape=', X_test.shape)
```

```python
def show_matrix(cm):
    labels = ['low', 'medium', 'high']
    group_counts = ["{0:0.0f}".format(value) for value in cm.flatten()]
    group_percentages = ["{0:.2%}".format(value) for value in cm.flatten() / np.sum(cm)]
    group = [f"{v1}\n{v2}" for v1, v2 in zip(group_counts, group_percentages)]
    group = np.asarray(group).reshape(cm.shape)
    fig, ax = plt.subplots(figsize=(8, 8))
    cax = ax.imshow(cm, cmap="Blues")
    fig.colorbar(cax)
    ax.set_xticks(np.arange(len(labels)))
    ax.set_yticks(np.arange(len(labels)))
    ax.set_xticklabels(labels)
    ax.set_yticklabels(labels)
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, group[i, j], ha='center', va='center', color='black')
    ax.set_title('Confusion Matrix with labels\n\n')
    ax.set_xlabel('\nPredicted Values')
    ax.set_ylabel('Actual Values ')
    plt.show()
    
def plot_precision_recall_curve():
    # Set plot title and labels
    plt.title('Precision-Recall Curves for All Classes')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.legend(loc='lower left')
    plt.grid(True)
    plt.show()

def plot_roc_curve(fpr, tpr, style='b-', label=None):
    plt.plot(fpr, tpr, style, lw=2)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve ', fontsize=20)
    plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Create an instance of RandomForestClassifier
forest_clf = RandomForestClassifier(n_estimators = 60, random_state = 42)

# Train the Classifier model
forest_clf.fit(X_train, y_train)

# Predict on training set using trained Classifier
y_pred_1 = forest_clf.predict(X_train)

train_acc = accuracy_score(y_train, y_pred_1)
print("Training accuracy: {:.4f}".format(train_acc))

# Create an unrealistic prediction array where all values are set to "low" (or any other single category)
y_train_alllow = np.full(len(y_train), "low")
print('Accuracy using all low values:', accuracy_score(y_train, y_train_alllow))
```
