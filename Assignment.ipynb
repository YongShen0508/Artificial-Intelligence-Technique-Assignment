# 2.0 Data Preprocessing

## 2.1 Split the dataset into training and testing set

```python
#Checking null value
data.isnull().sum()
```

Since there is no null value in the dataset, we can skip handling missing values.

```python
#Separating the output vector(Cardiovascular_risk(y)) from the input matrix
X = data.drop('Cardiovascular_risk(y)', axis = 1)
y = data['Cardiovascular_risk(y)']
```

```python
#Showing the shapes and types of X and y
print ('Shape of original dataset, train:', data.shape)
print ('X: shape=', X.shape, 'type=', type(X))
print ('y: shape=', y.shape, 'type=', type(y))
```

```python
#Split the dataset into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)

print ('full set shape=', X.shape)
print ('train shape=', X_train.shape)
print ('test shape=', X_test.shape)
```

```python
#Split the input matrix into numerical and categorical sets
X_train_num = X_train.drop(["Gender", "Family_history", "Alcohol", "Junk_food", "Snack", "Smoking", "Transportation", "TV", "Discipline"], axis=1)
X_train_cat = X_train[["Gender", "Family_history", "Alcohol", "Junk_food", "Snack", "Smoking", "Transportation", "TV", "Discipline"]]

print('X_train_num shape:', X_train_num.shape)
print('X_train_cat shape:', X_train_cat.shape)
```

There are 8 numerical attributes and 9 categorical attributes.

## 2.2 Preprocessing numerical data: Standardization

```python
#Standardizing the features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler(copy = False)
scaler.fit(X_train_num)       
X_train_num = scaler.transform(X_train_num)
```

```python
print('>>> Mean of all columns:\n', X_train_num.mean(axis=0))
print('\n>>> Std of all columns:\n', X_train_num.std(axis=0))
```

After standardization, the mean and standard deviation of X_train_num are approximately 0 and 1, respectively.

## 2.3 Preprocessing categorical Data: One-hot encoding

```python
#Encoding the categorical feature into its one-hot representation
from sklearn.preprocessing import OneHotEncoder

# Using OneHotEncoder to encode multiple categorical features
encoder = OneHotEncoder(sparse_output=False)
X_train_cat_tr = encoder.fit_transform(X_train_cat)
```

```python
print('Shape of X_train_cat_tr:', X_train_cat_tr.shape, '\n')
print(X_train_cat_tr, '\n')
```

Each of the 26 columns corresponds to one possible value of the categorical features.

```python
for feature, categories in zip(X_train_cat.columns, encoder.categories_):
    print(f"Feature '{feature}' possible values: {categories}")
```

These are the possible values for the transformed feature.

## 2.4 Finalizing the training set

```python
#Combining the transformed numerical and categorical sets
X_train_tr = np.hstack([X_train_num, X_train_cat_tr])
y_train = y_train.values;  #convert to numpy array

print('X_train_tr shape:', X_train_tr.shape)
print('y_train shape:', y_train.shape)
```

34 columns = 8 (Numerical features) + 26 (Possible value of the categorical features)
